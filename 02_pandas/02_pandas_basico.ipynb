Pandas permite trabajar con datos estructurados de forma intuitiva.

## Contenido

- Manejo de valores faltantes

```python
import pandas as pd

# Crear un DataFrame de ejemplo
data = {'gen': ['geneA', 'geneB', 'geneC'], 'expresion': [10.5, 4.2, 7.8], 'grupo': ['control', 'tratamiento', 'control']}
df = pd.DataFrame(data)
print(df)
print('\nMedia de expresión por grupo:')
print(df.groupby('grupo')['expresion'].mean())
```
