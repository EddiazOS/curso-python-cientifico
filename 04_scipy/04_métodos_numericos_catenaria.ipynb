## 1. Introducción al Problema

Imagina que queremos colgar una hamaca entre dos árboles. Sabemos dos cosas:

- La **distancia horizontal** entre los árboles es de **3 metros**.
- La **longitud total** de la hamaca (la tela extendida) es de **3.2 metros**.
- Los **ganchos de anclaje** están a **2 metros del suelo**.

**¿Preguntas que queremos responder?**

- ¿Qué forma toma exactamente la hamaca?
- ¿A qué altura sobre el suelo queda el punto más bajo de la hamaca (el "sag")?
- ¿Es esta hamaca adecuada para una persona sin tocar el suelo?

### El modelo físico: la Catenaria

La física nos dice que un cable o tela suspendida bajo su propio peso forma una curva llamada **catenaria**, cuya ecuación es:

$$y(x) = a \cosh\left(\frac{x}{a}\right)$$

donde $a > 0$ es un parámetro desconocido que controla qué tan "abierta" o "cerrada" es la curva.

Colocando el origen en el punto más bajo, la **longitud total** de la catenaria entre los puntos de anclaje en $x = -b$ y $x = b$ es:

$$L = 2a \sinh\left(\frac{b}{a}\right)$$

Nuestro objetivo es encontrar el valor de $a$ que satisfaga esta ecuación para nuestros valores conocidos de $L$ y $b$.

---

## 2. Planteamiento de la Ecuación a Resolver

No podemos despejar $a$ algebraicamente de la ecuación de longitud. Por lo tanto, reescribimos el problema como la búsqueda de la **raíz** de una función $F(a) = 0$:

$$F(a) = 2a \sinh\left(\frac{b}{a}\right) - L = 0$$

Usaremos el **Método de Newton-Raphson** para encontrar esta raíz numéricamente.

### El Método de Newton-Raphson

El método construye una secuencia de aproximaciones $a_0, a_1, a_2, \ldots$ usando la regla de actualización:

$$a_{i+1} = a_i - \frac{F(a_i)}{F'(a_i)}$$

En lugar de calcular $F'(a)$ de forma analítica, usamos una **derivada numérica** por diferencias hacia adelante con un paso pequeño $h$:

$$F'(a) \approx \frac{F(a + h) - F(a)}{h}$$

Esto nos permite aplicar Newton-Raphson a cualquier función, sin necesidad de derivar analíticamente.

## 6. ¿Es Adecuada la Longitud de la Hamaca?

Con el parámetro $a$ ya encontrado, podemos calcular la **altura del punto más bajo** de la hamaca sobre el suelo.

### Deducción

La catenaria pura $y(x) = a\cosh(x/a)$ tiene:

- **Punto más bajo** (en $x=0$): $y_{\min} = a\cosh(0) = a$
- **Altura en los extremos** (en $x = \pm b$): $y_{\text{ext}} = a\cosh(b/a)$

El **sag** (cuánto cuelga la hamaca respecto a los anclajes) es:

$$\text{sag} = y_{\text{ext}} - y_{\min} = a\cosh\left(\frac{b}{a}\right) - a$$

Y la **altura del punto más bajo sobre el suelo** es:

$$H_{\min} = H_{\text{anclaje}} - \text{sag}$$

## 📚 Resumen de Conceptos

| Concepto | Descripción |
|----------|-------------|
| **Catenaria** | Curva que forma un cable flexible suspendido: $y = a\cosh(x/a)$ |
| **Ecuación de longitud** | $L = 2a\sinh(b/a)$ — relaciona la longitud con el parámetro $a$ y el vano $b$ |
| **Búsqueda de raíces** | Reformular el problema como $F(a) = 0$ para resolverlo numéricamente |
| **Derivada numérica** | $F'(a) \approx (F(a+h) - F(a))/h$ — evita derivar analíticamente |
| **Newton-Raphson** | Método iterativo: $a_{i+1} = a_i - F(a_i)/F'(a_i)$ — converge muy rápido |
| **Sag** | Distancia vertical entre los anclajes y el punto más bajo de la hamaca |
