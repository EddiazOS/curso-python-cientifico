```python
from scipy import stats
import numpy as np

# Prueba t de dos muestras independientes
grupo_control = np.array([5.1, 4.9, 5.3, 5.0, 5.2])
grupo_tratamiento = np.array([6.2, 6.5, 6.1, 6.8, 6.3])

t_stat, p_valor = stats.ttest_ind(grupo_control, grupo_tratamiento)
print(f'Estadístico t: {t_stat:.3f}')
print(f'Valor p: {p_valor:.4f}')
if p_valor < 0.05:
    print('Diferencia estadísticamente significativa (p < 0.05)')
```
