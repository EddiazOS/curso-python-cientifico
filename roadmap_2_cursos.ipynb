## 📋 Curso 1 — Fundamentos (8 semanas, 16 sesiones)

| Semana | Sesión | Módulo | Temas |
|--------|--------|--------|-------|
| 1 | 1-2 | `00_introduccion` | Sintaxis básica, control de flujo, funciones |
| 2 | 3-4 | `01_numpy` | Arrays, indexación, operaciones vectorizadas |
| 3 | 5-6 | `01_numpy` | Álgebra lineal básica, producto matricial, normas |
| 4 | 7-8 | `02_pandas` | DataFrames, selección, filtrado, groupby |
| 5 | 9-10 | `03_matplotlib` | Gráficos 1D/2D, histogramas, scatter, error bars |
| 6 | 11-12 | `04_scipy` | Búsqueda de raíces (bisección, Newton-Raphson), catenaria |
| 7 | 13-14 | `04_scipy` | Ajuste de curvas (curve_fit), optimización básica |
| 8 | 15-16 | `proyecto_1` | Mini-proyecto: análisis de datos + visualización |

**Meta Curso 1:** Manipular arrays, dataframes, graficar, resolver ecuaciones no lineales, ajustar modelos simples.

## 🧬 Curso 2 — Bio/Quimioinformática (6 semanas, 12 sesiones)

| Semana | Sesión | Módulo | Temas |

| 1 | 1-2 | `bioinformatica` | Biopython, secuencias FASTA, GC%, ORFs |

**Meta Curso 2:** Análisis real de microbioma, descriptores químicos, modelos biológicos con bibliotecas especializadas.
