```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 2 * np.pi, 100)
y = np.sin(x)

plt.figure(figsize=(8, 4))
plt.plot(x, y, color='steelblue', linewidth=2)
plt.title('Función Seno')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.tight_layout()
plt.show()
```
