## 2. Variables y tipos de datos

Una **variable** es un nombre que le asignamos a un valor para poder usarlo después. Piensa en ella como una etiqueta pegada a una caja que guarda información.

```python
nombre_variable = valor
```

Python reconoce automáticamente el **tipo** del valor que guardas. Los tipos fundamentales son:

| Tipo | Descripción | Ejemplo |
|------|-------------|---------|
| `int` | Número entero | `42`, `-7` |
| `float` | Número decimal | `3.14`, `-0.5` |
| `str` | Cadena de texto | `"Homo sapiens"` |
| `bool` | Valor lógico | `True`, `False` |
| `None` | Ausencia de valor | `None` |

> Usa `type(variable)` para saber qué tipo tiene cualquier variable.

````python
# Biología: descripción de un organismo
species_name = 'Escherichia coli'   # str
genome_size_mb = 4.6                # float (megabases)
num_genes = 4377                    # int
is_pathogenic = False               # bool
plasmid_sequence = None             # aún no disponible

print(species_name, '->', type(species_name))
print(genome_size_mb, '->', type(genome_size_mb))
print(num_genes, '->', type(num_genes))
print(is_pathogenic, '->', type(is_pathogenic))
print(plasmid_sequence, '->', type(plasmid_sequence))

# Química: propiedades de un compuesto
compound_name = 'Glucosa'
molecular_weight = 180.16           # g/mol
num_carbons = 6
is_reducing_sugar = True

print(f'\nCompuesto: {compound_name}')
print(f'Peso molecular: {molecular_weight} g/mol')
print(f'Carbonos: {num_carbons}')
print(f'Es azúcar reductor: {is_reducing_sugar}')
````

## 3. Operaciones básicas

Python puede usarse como una calculadora científica muy poderosa. Los operadores aritméticos principales son:

| Operador | Operación | Ejemplo | Resultado |
|----------|-----------|---------|-----------|
| `+` | Suma | `3 + 2` | `5` |
| `-` | Resta | `5 - 1` | `4` |
| `*` | Multiplicación | `4 * 3` | `12` |
| `/` | División | `7 / 2` | `3.5` |
| `//` | División entera | `7 // 2` | `3` |
| `%` | Módulo (residuo) | `7 % 2` | `1` |
| `**` | Potencia | `2 ** 8` | `256` |

Los operadores de **comparación** devuelven `True` o `False`:
`==`, `!=`, `<`, `>`, `<=`, `>=`

````python
# Dilución: C1 * V1 = C2 * V2  →  calcular V1
C1 = 1000.0   # concentración stock (µg/mL)
C2 = 50.0     # concentración deseada (µg/mL)
V2 = 10.0     # volumen final (mL)
V1 = (C2 * V2) / C1
print(f'Volumen a tomar del stock: {V1} mL')

# Ley de Beer-Lambert: A = epsilon * c * l
epsilon = 6500   # coeficiente de extinción molar (L/mol·cm)
c = 0.002        # concentración (mol/L)
l = 1.0          # longitud de paso (cm)
A = epsilon * c * l
print(f'Absorbancia: {A}')

# Conversión de temperatura: °C a K
temp_c = 37.0
temp_k = temp_c + 273.15
print(f'{temp_c}°C = {temp_k} K')

# Operadores de comparación
print(f'\n¿A > 1? {A > 1}')
print(f'¿temp_c == 37.0? {temp_c == 37.0}')
print(f'¿V1 != V2? {V1 != V2}')
````

## 4. Cadenas de texto (`str`)

Las cadenas de texto son fundamentales en bioinformática: secuencias de ADN, ARN y proteínas son simplemente `str`.

Operaciones útiles con strings:
- **Concatenación**: `'AT' + 'CG'` → `'ATCG'`
- **Repetición**: `'AT' * 3` → `'ATATАТ'`
- **Indexación**: `seq[0]` devuelve el primer carácter
- **Slicing**: `seq[0:3]` devuelve los primeros 3 caracteres
- **f-strings**: forma moderna de insertar variables en texto

Métodos importantes: `.upper()`, `.lower()`, `.replace()`, `.count()`, `.strip()`, `len()`

````python
# Secuencia de ADN como string
dna = 'atgcgcttaaccggta'

print('Original:', dna)
print('Mayúsculas:', dna.upper())
print('Longitud:', len(dna))
print('Primer nucleótido:', dna[0])
print('Últimos 4 nucleótidos:', dna[-4:])
print('Primeros 6 nucleótidos:', dna[0:6])

# Contar bases
dna = dna.upper()
print(f'\nConteo de bases en {dna}:')
print(f'  A: {dna.count("A")}')
print(f'  T: {dna.count("T")}')
print(f'  G: {dna.count("G")}')
print(f'  C: {dna.count("C")}')

# Reemplazar T por U (transcripción ADN → ARN)
rna = dna.replace('T', 'U')
print(f'\nADN: {dna}')
print(f'ARN: {rna}')

# Concatenar dos fragmentos
fragment_1 = 'ATGCGA'
fragment_2 = 'TTACGG'
full_seq = fragment_1 + fragment_2
print(f'\nSecuencia completa: {full_seq} (longitud: {len(full_seq)})')
````

## 5. Estructuras de datos

Cuando necesitas guardar **múltiples valores**, Python ofrece cuatro estructuras fundamentales:

| Estructura | Sintaxis | Ordenada | Modificable | Duplicados |
|------------|----------|----------|-------------|------------|
| Lista | `[...]` | ✅ | ✅ | ✅ |
| Tupla | `(...)` | ✅ | ❌ | ✅ |
| Diccionario | `{clave: valor}` | ✅ | ✅ | ❌ (claves) |
| Conjunto | `set(...)` | ❌ | ✅ | ❌ |


````python
# LISTA: muestras de ADN recolectadas
sequences = ['ATCGGTA', 'GCTATCG', 'TTAACCG', 'ATCGGTA']
print('Secuencias:', sequences)
print('Primera:', sequences[0])
print('Última:', sequences[-1])
sequences.append('CGTAGCT')
print('Después de agregar:', sequences)

# TUPLA: coordenadas fijas de una muestra de campo (lat, lon, altitud)
sample_location = (-3.745, -73.253, 120)
lat, lon, alt = sample_location
print(f'\nUbicación — Lat: {lat}, Lon: {lon}, Alt: {alt} m')

# DICCIONARIO: ficha de un compuesto químico
compound = {
    'name': 'Ácido ascórbico',
    'formula': 'C6H8O6',
    'molecular_weight': 176.12,
    'soluble_in_water': True
}
print(f"\nCompuesto: {compound['name']} ({compound['formula']})")
print(f"PM: {compound['molecular_weight']} g/mol")

# CONJUNTO: genes detectados en dos muestras
genes_A = {'BRCA1', 'TP53', 'MYC', 'EGFR'}
genes_B = {'TP53', 'EGFR', 'KRAS', 'VEGF'}
print('\nGenes en común:', genes_A & genes_B)
print('Genes únicos de A:', genes_A - genes_B)
print('Unión total:', genes_A | genes_B)
````

## 6. Estructuras de control

Las estructuras de control permiten que tu programa **tome decisiones** y **repita operaciones**.

### `if / elif / else` — Condicionales
Ejecuta un bloque de código solo si se cumple una condición.

### `for` — Bucle sobre una secuencia
Itera sobre cada elemento de una lista, tupla, string, etc.

### `while` — Bucle mientras se cumpla una condición
Repite un bloque mientras una condición sea `True`. Útil cuando no sabes de antemano cuántas veces repetir.

> ⚠️ En Python, la **indentación** (espacios al inicio de línea) es obligatoria y define los bloques de código.

````python
# IF/ELIF/ELSE: clasificar pH de una solución
ph = 6.2

if ph < 7:
    solution_type = 'ácida'
elif ph == 7:
    solution_type = 'neutra'
else:
    solution_type = 'básica'

print(f'pH = {ph} → Solución {solution_type}')

# FOR: recorrer una lista de compuestos e imprimir sus propiedades
compounds = [
    {'name': 'Agua',    'formula': 'H2O',    'mw': 18.02},
    {'name': 'Glucosa', 'formula': 'C6H12O6','mw': 180.16},
    {'name': 'NaCl',    'formula': 'NaCl',   'mw': 58.44},
]

print('\nCompuesto          | Fórmula    | PM (g/mol)')
print('-' * 44)
for c in compounds:
    print(f"{c['name']:<18} | {c['formula']:<10} | {c['mw']}")

# WHILE: simular diluciones seriadas
concentration = 1000.0
dilution_factor = 10
step = 0

print('\nDiluciones seriadas (factor 1:10):')
while concentration >= 1:
    print(f'  Paso {step}: {concentration:.2f} µg/mL')
    concentration /= dilution_factor
    step += 1
````

## 7. Funciones

Una **función** es un bloque de código reutilizable que recibe entradas (**parámetros**) y devuelve una salida (`return`). Evita repetir código y hace tus análisis más reproducibles.

```python
def nombre_funcion(parametro1, parametro2='valor_por_defecto'):
    """Descripción de la función (docstring)."""
    # cuerpo
    return resultado
```

El **docstring** (texto entre triple comillas) documenta qué hace la función. Es una buena práctica siempre incluirlo.

````python
# Función 1: contenido GC de una secuencia de ADN
def gc_content(sequence):
    """
    Calcula el porcentaje de GC de una secuencia de ADN.
    Args:
        sequence (str): Secuencia de ADN (A, T, C, G).
    Returns:
        float: Porcentaje de GC (0 a 100).
    """
    sequence = sequence.upper()
    gc = sequence.count('G') + sequence.count('C')
    return (gc / len(sequence)) * 100

print(f"GC de 'ATGCGCTTAACC': {gc_content('ATGCGCTTAACC'):.1f}%")
print(f"GC de 'GGGCCCATGCAT': {gc_content('GGGCCCATGCAT'):.1f}%")

# Función 2: conversión de temperatura
def celsius_to_kelvin(temp_c):
    """Convierte grados Celsius a Kelvin."""    return temp_c + 273.15

def celsius_to_fahrenheit(temp_c):
    """Convierte grados Celsius a Fahrenheit."""    return (temp_c * 9/5) + 32

temp = 37.0
print(f'\n{temp}°C = {celsius_to_kelvin(temp)} K')
print(f'{temp}°C = {celsius_to_fahrenheit(temp):.1f}°F')

# Función 3: concentración molar
def molar_concentration(mass_g, mw, volume_L):
    """
    Calcula la concentración molar de una solución.
    Args:
        mass_g (float): Masa del soluto en gramos.
        mw (float): Peso molecular en g/mol.
        volume_L (float): Volumen de la solución en litros.
    Returns:
        float: Concentración en mol/L.
    """
    moles = mass_g / mw
    return moles / volume_L

conc = molar_concentration(mass_g=5.0, mw=180.16, volume_L=0.5)
print(f'\nConcentración de glucosa: {conc:.4f} M')
````

## 8. Comprensiones de listas

Las **list comprehensions** son una forma compacta y elegante de crear listas a partir de una secuencia existente:

```python
nueva_lista = [expresión for elemento in secuencia if condición]
```

Son equivalentes a un bucle `for`, pero en una sola línea. Son muy usadas en análisis de datos científicos para transformar y filtrar colecciones.

````python
# Ejemplo 1: limpiar y estandarizar secuencias
raw_sequences = ['atcggta', 'gctatcg', 'ttaaccg']
clean_sequences = [seq.upper() for seq in raw_sequences]
print('Secuencias limpias:', clean_sequences)

# Ejemplo 2: calcular GC de múltiples secuencias a la vez
gc_values = [gc_content(seq) for seq in clean_sequences]
print('Contenidos GC (%):', gc_values)

# Ejemplo 3: filtrar solo secuencias con GC > 50%
high_gc = [seq for seq in clean_sequences if gc_content(seq) > 50]
print('Secuencias con GC > 50%:', high_gc)

# Ejemplo 4: química — nombres de compuestos solubles en agua
all_compounds = [
    {'name': 'NaCl',    'soluble': True,  'mw': 58.44},
    {'name': 'Benceno', 'soluble': False, 'mw': 78.11},
    {'name': 'Glucosa', 'soluble': True,  'mw': 180.16},
    {'name': 'Etanol',  'soluble': True,  'mw': 46.07},
]
soluble = [c['name'] for c in all_compounds if c['soluble']]
print('\nCompuestos solubles en agua:', soluble)

# Equivalente con bucle for (para comparar)
soluble_loop = []
for c in all_compounds:
    if c['soluble']:
        soluble_loop.append(c['name'])
print('Mismo resultado con for:', soluble_loop)
````

## 9. Manejo de archivos

En ciencia trabajamos constantemente con archivos: datos de experimentos, secuencias, resultados. Python puede leer y escribir archivos `.txt` y `.csv` directamente.

```python
with open('archivo.txt', 'w') as f:   # 'w' = escribir, 'r' = leer, 'a' = agregar
    f.write('contenido')
```

Usamos el bloque `with` para garantizar que el archivo se cierre correctamente al terminar, incluso si ocurre un error.

> En el módulo de **Pandas** aprenderemos formas más poderosas de trabajar con archivos de datos tabulares (`.csv`, `.xlsx`).

````python
import os

# ESCRITURA: guardar resultados de análisis de secuencias
results = [
    ('ATCGGTA',  gc_content('ATCGGTA')),
    ('GCTATCG',  gc_content('GCTATCG')),
    ('TTAACCG',  gc_content('TTAACCG')),
    ('GGGCCCATG',gc_content('GGGCCCATG')),
]

with open('gc_results.txt', 'w') as f:
    f.write('sequence\tgc_content(%)\n')
    for seq, gc in results:
        f.write(f'{seq}\t{gc:.2f}\n')

print("Archivo 'gc_results.txt' creado exitosamente.")

# LECTURA: leer el archivo generado
print('\nContenido del archivo:')
print('-' * 30)
with open('gc_results.txt', 'r') as f:
    for line in f:
        print(line.strip())

# Limpieza del archivo de ejemplo
os.remove('gc_results.txt')
print('\nArchivo eliminado tras la demostración.')
````
